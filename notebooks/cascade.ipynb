```python
import pickle
from sklearn.metrics import log_loss, roc_auc_score
import numpy as np
from cpx_model.inference import compute_confusion_matrix
import torch
```

## Open Probabilities File

```python
cpx_prob_dir = 'cpx_model/inference_logs/probabilities_20251116-214238.pkl'
bert_prob_dir = 'bert_routing/inference_logs/probabilities_20251117-151544.pkl'

with open(bert_prob_dir, 'rb') as f:
    bert_validation_data = pickle.load(f)
bert_probabilities = bert_validation_data['probabilities']
labels = bert_validation_data['labels']

with open(cpx_prob_dir, 'rb') as f:
    cpx_validation_data = pickle.load(f)

cpx_probabilities = cpx_validation_data['probabilities']
```

```python
#BERT roc_auc_score
bert_roc_auc_score = roc_auc_score(labels, bert_probabilities)
print(f'bert roc_auc_score: {bert_roc_auc_score}')

#CPX roc_auc_score
cpx_roc_auc_score = roc_auc_score(labels, cpx_probabilities)
print(f'cpx roc_auc_score: {cpx_roc_auc_score}')
```

```python
thresholds = list(np.arange(0.01, 0.99, 0.01))
threshold = 0.15
print(f'total samples: {len(labels)}')
# Compute confusion matrix for cpx
cpx_confusion_matrix = compute_confusion_matrix(cpx_probabilities, labels, threshold)
print(f'cpx total negative: {cpx_confusion_matrix["TN"] + cpx_confusion_matrix["FN"]}')
# Negative precision:
try:
    print(f'cpx negative precision: {cpx_confusion_matrix["TN"]/(cpx_confusion_matrix["TN"] + cpx_confusion_matrix["FN"])}')
except ZeroDivisionError:
    print(f'No simple labels detected')

# Compute confusion matrix for bert
bert_confusion_matrix = compute_confusion_matrix(bert_probabilities, labels, threshold)
print(f'bert total negative: {bert_confusion_matrix["TN"] + bert_confusion_matrix["FN"]}')
# Negative precision:
try:
    print(f'bert negative precision: {bert_confusion_matrix["TN"]/(bert_confusion_matrix["TN"] + bert_confusion_matrix["FN"])}')
except ZeroDivisionError:
    print(f'No simple labels detected')

```
